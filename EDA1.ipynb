```python
d3 = df["DS"].mean()
df["DS"] = df["DS"].fillna(value=d3)
```

```python
d4 = df["DP"].mean()
df["DP"] = df["DP"].fillna(value=d4)
```

```python
d5 = df["MLTV"].mean()
df["MLTV"] = df["MLTV"].fillna(value=d5)
```

```python
d6 = df["ALTV"].mean()
df["ALTV"] = df["ALTV"].fillna(value=d6)
```

```python
d7 = df["Width"].mean()
df["Width"] = df["Width"].fillna(value=d7)
```
