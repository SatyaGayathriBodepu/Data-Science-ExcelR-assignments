```python
# standardization
```

```python
X = df.drop("Type",axis=1)
Y = df["Type"]
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
```

```python
SS_X = SS.fit_transform(X)
SS_X = pd.DataFrame(SS_X)
```

```python
# data partition
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(SS_X,Y,test_size=0.3,random_state=42)
```

```python
# random forest
```

```python
from sklearn.metrics import accuracy_score,precision_score, recall_score, f1_score,log_loss
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RF_model = RandomForestClassifier(n_estimators=100,random_state=42,max_samples=0.6,
                                       max_features=0.7)
RF_model.fit(X_train,Y_train)

# Predictions
y_pred_train_RF = RF_model.predict(X_train)
y_pred_test_RF = RF_model.predict(X_test)

# Probabilities
y_pred_train_proba_RF = RF_model.predict_proba(X_train)
y_pred_test_proba_RF = RF_model.predict_proba(X_test)

# Metrics
training_accuracy_RF = accuracy_score(Y_train, y_pred_train_RF)
testing_accuracy_RF = accuracy_score(Y_test, y_pred_test_RF)

training_loss_RF = log_loss(Y_train, y_pred_train_proba_RF)
testing_loss_RF = log_loss(Y_test, y_pred_test_proba_RF)

print("RF - Training Accuracy:", training_accuracy_RF)
print("RF - Testing Accuracy:", testing_accuracy_RF)
print("RF - Training Log Loss:", training_loss_RF)
print("RF - Testing Log Loss:", testing_loss_RF)
```

```python
precision = precision_score(Y_test, y_pred_test_RF, average='weighted')
recall = recall_score(Y_test, y_pred_test_RF, average='weighted')
f1 = f1_score(Y_test, y_pred_test_RF, average='weighted')

print("Test Precision:", precision)
print("Test Recall   :", recall)
print("Test F1-score :", f1)
```

```python
# bagging
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
bagging_model = BaggingClassifier(estimator=DecisionTreeClassifier(criterion='gini'),n_estimators=100,
                                   random_state=42,max_samples=0.6,max_features=0.7)
bagging_model.fit(X_train,Y_train)

# Predictions
y_pred_train_bagging = bagging_model.predict(X_train)
y_pred_test_bagging = bagging_model.predict(X_test)

# Probabilities
y_pred_train_proba_bagging = bagging_model.predict_proba(X_train)
y_pred_test_proba_bagging = bagging_model.predict_proba(X_test)

# Accuracy
training_accuracy_bagging = accuracy_score(Y_train, y_pred_train_bagging)
testing_accuracy_bagging = accuracy_score(Y_test, y_pred_test_bagging)

# Log Loss (ONLY with probabilities)
training_loss_bagging = log_loss(Y_train, y_pred_train_proba_bagging)
testing_loss_bagging = log_loss(Y_test, y_pred_test_proba_bagging)

print("Bagging - Training Accuracy:", training_accuracy_bagging)
print("Bagging - Testing Accuracy:", testing_accuracy_bagging)
print("Bagging - Training Log Loss:", training_loss_bagging)
print("Bagging - Testing Log Loss:", testing_loss_bagging)
```

```python
# Boosting
```

```python
from sklearn.ensemble import GradientBoostingClassifier
GBM=GradientBoostingClassifier(n_estimators=100,max_features=0.7,learning_rate=0.1)
```

```python
GBM.fit(X_train,Y_train)

#Predictions
y_pred_train_gb = GBM.predict(X_train)
y_pred_test_gb = GBM.predict(X_test)

# Probabilities
y_pred_train_proba_gb = GBM.predict_proba(X_train)
y_pred_test_proba_gb = GBM.predict_proba(X_test)

# Accuracy
training_accuracy_gb = accuracy_score(Y_train, y_pred_train_gb)
testing_accuracy_gb = accuracy_score(Y_test, y_pred_test_gb)

# Log Loss (ONLY with probabilities)
training_loss_gb = log_loss(Y_train, y_pred_train_proba_gb)
testing_loss_gb = log_loss(Y_test, y_pred_test_proba_gb)

print("GB - Training Accuracy:", training_accuracy_gb)
print("GB - Testing Accuracy:", testing_accuracy_gb)
print("GB - Training Log Loss:", training_loss_gb)
print("GB - Testing Log Loss:", testing_loss_gb)
```
