```python
# data partition
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5)
```

```python
# kvalue cross validation
```

```python
k_values = list(range(5, 18, 2))
training_accuracy_list = []
testing_accuracy_list = []
```

```python
for k in k_values:
    training_accuracy = []
    testing_accuracy = []

    for i in range(1, 50):  # number of repetitions for shuffle-split
        X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=i)
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train, Y_train)
        y_pred_train = knn.predict(X_train)
        y_pred_test = knn.predict(X_test)
        training_accuracy.append(accuracy_score(Y_train, y_pred_train))
        testing_accuracy.append(accuracy_score(Y_test, y_pred_test))

    training_accuracy_list.append(np.mean(training_accuracy).round(2))
    testing_accuracy_list.append(np.mean(testing_accuracy).round(2))
print("Training Accuracies:",training_accuracy_list)
print("Testing Accuracies:",testing_accuracy_list)
```

```python
print(len(k_values), len(training_accuracy_list), len(testing_accuracy_list))
```

```python
import matplotlib.pyplot as plt

plt.plot(k_values, training_accuracy_list, marker='o', color='red', label='Training Accuracy')
plt.plot(k_values, testing_accuracy_list, marker='o', color='green', label='Testing Accuracy')
plt.xlabel("K value")
plt.ylabel("Accuracy")
plt.title("Training vs Testing Accuracy for different K")
plt.legend()
plt.show()
```

```python
# visualize decision boundaries of classifier
```

```python
from matplotlib.colors import ListedColormap
```

```python
X_vis = X[['hair','legs']].values
y_vis = Y.values
```

```python
knn_vis = KNeighborsClassifier(n_neighbors=5)
knn_vis.fit(X_vis, y_vis)
```

```python
x_min, x_max = X_vis[:,0].min() - 1, X_vis[:,0].max() + 1
y_min, y_max = X_vis[:,1].min() - 1, X_vis[:,1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))
```

```python
Z = knn_vis.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
```

```python
plt.figure(figsize=(8,6))
cmap = ListedColormap(['red', 'green', 'blue', 'yellow', 'cyan', 'magenta', 'orange'])
plt.contourf(xx, yy, Z, alpha=0.3, cmap=cmap)
plt.scatter(X_vis[:,0], X_vis[:,1], c=y_vis, edgecolor='k', cmap=cmap)
plt.xlabel('Hair')
plt.ylabel('Legs')
plt.title('KNN Decision Boundaries (Hair vs Legs)')
plt.show()
```
